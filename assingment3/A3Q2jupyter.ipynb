```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lorenz(x, y, z, σ=10, r=28, b=2.667):
    """
    Given:
       x, y, z: a point of interest in three dimensional space
       σ, r, b: parameters defining the lorenz attractor
    Returns:
       x_dot, y_dot, z_dot: values of the lorenz attractor's partial
           derivatives at the point x, y, z
    """
    x_dot = σ*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot
```

```python
dt = 0.01
num_steps = 3000

# Need one more for the initial values
xσ = np.empty(num_steps + 1)
yσ = np.empty(num_steps + 1)
zσ = np.empty(num_steps + 1)
```

```python
# Set w0 values
# Need one more for the initial values
xσ1 = np.empty(num_steps)
yσ1 = np.empty(num_steps)
zσ1 = np.empty(num_steps)

xσ[0], yσ[0], zσ[0] = (0., 1., 0.)

# Step through "time", calculating the partial derivatives at the current point
# and using them to estimate the next point
for i in range(num_steps):
    x_dot, y_dot, z_dot = lorenz(xσ[i], yσ[i], zσ[i])
    xσ[i + 1] = xσ[i] + (x_dot * dt)
    yσ[i + 1] = yσ[i] + (y_dot * dt)
    zσ[i + 1] = zσ[i] + (z_dot * dt)
```

```python
# Plot
ax = plt.figure().add_subplot(projection='3d')

ax.plot(xσ, yσ, zσ, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```
